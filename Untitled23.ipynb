#**1. Data Preparation**

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris['data']
y = iris['target']
names = iris['target_names']
feature_names = iris['feature_names']

# Scale the data to have mean 0 and variance 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=2)

# Print data shapes
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
```

#**2. Neural Network Model Definition**

```python
# Import PyTorch libraries
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable

# Convert training and testing data to PyTorch variables
X_train = Variable(torch.from_numpy(X_train)).float()
y_train = Variable(torch.from_numpy(y_train)).long()
X_test = Variable(torch.from_numpy(X_test)).float()
y_test = Variable(torch.from_numpy(y_test)).long()

# Define the neural network model
class Model(nn.Module):
    def __init__(self, input_dim):
        super(Model, self).__init__()
        self.layer1 = nn.Linear(input_dim, 50)
        self.layer2 = nn.Linear(50, 50)
        self.layer3 = nn.Linear(50, 3)

    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.softmax(self.layer3(x), dim=1)
        return x

# Instantiate the model
model = Model(X_train.shape[1])

# Print the model architecture
print(model)
```

#**3. Training Setup and Optimizer Configuration**

```python
# Import optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.CrossEntropyLoss()

# Set number of epochs
EPOCHS = 2000

# Initialize lists to store loss and accuracy values
loss_list_train = np.zeros((EPOCHS,))
loss_list_val = np.zeros((EPOCHS,))
accuracy_list = np.zeros((EPOCHS,))
```

#**4. Model Training with Validation Loss**

```python
# Import tqdm for progress bar
import tqdm

# Training loop
for epoch in tqdm.trange(EPOCHS):
    # Forward pass for training set
    y_pred_train = model(X_train)
    loss_train = loss_fn(y_pred_train, y_train)
    loss_list_train[epoch] = loss_train.item()

    # Backpropagation and optimization
    optimizer.zero_grad()
    loss_train.backward()
    optimizer.step()

    # Evaluate on validation set (test set)
    with torch.no_grad():
        y_pred_val = model(X_test)
        loss_val = loss_fn(y_pred_val, y_test)
        loss_list_val[epoch] = loss_val.item()
        correct = (torch.argmax(y_pred_val, dim=1) == y_test).type(torch.FloatTensor)
        accuracy_list[epoch] = correct.mean()

# Print final training and validation loss
print(f"Final Training Loss: {loss_list_train[-1]}")
print(f"Final Validation Loss: {loss_list_val[-1]}")
```

#**5. Plot Accuracy and Loss during Training**

```python
# Plot accuracy and loss
fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 6), sharex=True)

# Plot accuracy over epochs
ax1.plot(accuracy_list, label="Validation Accuracy")
ax1.set_ylabel("Validation Accuracy")
ax1.legend()

# Plot loss for both training and validation
ax2.plot(loss_list_train, label="Training Loss")
ax2.plot(loss_list_val, label="Validation Loss")
ax2.set_ylabel("Loss")
ax2.set_xlabel("Epochs")
ax2.legend()

plt.show()
```

#**6. Confusion Matrix and Classification Report**

```python
# Import metrics
from sklearn.metrics import confusion_matrix, classification_report

# Get predicted classes for the test set
y_pred_class = torch.argmax(model(X_test), dim=1)

# Compute confusion matrix and classification report
conf_matrix = confusion_matrix(y_test, y_pred_class)
class_report = classification_report(y_test, y_pred_class, target_names=names)

# Print confusion matrix and classification report
print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
```

#**7. ROC Curve and AUC for Each Class**

```python
# Import additional metrics
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import OneHotEncoder

# One hot encode the target variable for ROC calculation
enc = OneHotEncoder()
Y_onehot = enc.fit_transform(y_test[:, np.newaxis]).toarray()

# Plot ROC curves for each class
plt.figure(figsize=(10, 10))
plt.plot([0, 1], [0, 1], 'k--')

# Compute and plot ROC for each class
with torch.no_grad():
    y_pred = model(X_test).numpy()
    for i in range(3):  # 3 classes in the Iris dataset
        fpr, tpr, _ = roc_curve(Y_onehot[:, i], y_pred[:, i])
        plt.plot(fpr, tpr, label='Class {} AUC = {:.3f}'.format(names[i], auc(fpr, tpr)))

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Class')
plt.legend()
plt.show()
```

#**Print final test accuracy**

```python
# Print final test accuracy
final_test_accuracy = accuracy_list[-1] * 100
print(f"Final test accuracy: {final_test_accuracy:.2f}%")
```
