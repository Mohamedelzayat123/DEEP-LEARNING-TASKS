#**Part 1: Import Necessary Packages**

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler  # For scaling features
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report, roc_curve, auc
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import RocCurveDisplay

```

#**Part 2: Set Random Seed**

```python

RANDOM_SEED = 40
tf.random.set_seed(RANDOM_SEED)

```

#**Part 3: Load Dataset**

```python

# Read the dataset from the CSV file
df = pd.read_csv("heart.csv")  # Ensure your CSV file is named "heart.csv"

# Display the first few rows of the dataset
df.head()
```

#**Part 4: Prepare Data**

```python

# Split the data into independent variables (X) and dependent variable (y)
X = df.drop(columns=["HeartDiseaseorAttack"]).values  # All columns except target
y = df["HeartDiseaseorAttack"].values  # Target column

# Scale the features using StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=RANDOM_SEED)

# Determine the number of input features
n_features = X_train.shape[1]  # Number of features in the training data
print("Number of features:", n_features)
```

#**Part 5: Define the Model**

```python


# Create a sequential model
model = Sequential()
# Add the first hidden layer with 26 nodes and ReLU activation
model.add(Dense(26, activation='relu', input_shape=(n_features,)))
# Add the second hidden layer with 13 nodes and ReLU activation
model.add(Dense(13, activation='relu'))
# Add the output layer with 1 node and sigmoid activation for binary classification
model.add(Dense(1, activation='sigmoid'))

```

#**Part 6: Compile the Model**

```python

# Compile the model with Adam optimizer, binary cross-entropy loss, and accuracy metric
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

#**Part 7: Train the Model**

```python

# Fit the model to the training data for 150 epochs with a batch size of 32 and show training progress
history = model.fit(X_train, y_train, epochs=150, batch_size=32, verbose=1, validation_split=0.2)

```

#**Part 8: Make Predictions**

```python

# Perform predictions on the test data
y_pred = model.predict(X_test)

# Round the predictions to get binary output (0 or 1)
y_pred1 = y_pred.round()
```

#**Part 9: Evaluate the Model**

```python

# Calculate the accuracy score
acc = accuracy_score(y_test, y_pred1)
print('Test Accuracy: %.3f' % acc)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred1)
print('Confusion Matrix:\n', conf_matrix)

# Display classification report
print('Classification Report:\n', classification_report(y_test, y_pred1))

# Calculate the F1 score with weighted averaging
f1 = f1_score(y_test, y_pred1, average='weighted')
print('F1 Score: %.3f' % f1)

# ROC-AUC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
print('ROC AUC: %.3f' % roc_auc)

# Plot ROC-AUC curve
plt.figure()
plt.plot(fpr, tpr, color='blue', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()

```

#**Part 10: Visualize Training History**

```python

# Plot training & validation loss values
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()


# Plot training & validation accuracy values
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.show()
```
